```python
from langserve.client import RemoteRunnable

rag_app = RemoteRunnable("http://127.0.0.1:8000/rag-conversation")
question = "概括一下元宇宙的几个发展阶段?"
answer = rag_app.invoke(
    {
        "question": question,
        "chat_history": [],
    }
)
```

```python
answer
```

```python
chat_history = [(question, answer)]
answer = rag_app.invoke(
    {
        "question": "举例解释元宇宙的本质",
        "chat_history": chat_history,
    }
)
```

```python
answer
```
